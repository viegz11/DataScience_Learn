# Python Fundamentals for Data Science


## Variables & Data Types

```python
name="Vignesh"
age=21
cgpa=7.8
is_student=True
print(type(name), type(age), type(cgpa), type(is_student))
```

## Operators

```python
a=10
b=3
print(a+b,a-b,a*b,a/b,a%b,a**b)
print(a>b, a==b)
print((a>5) and (b<5))
```

## Conditional Statements

```python
marks=85
if marks>=90:
    print("Grade A")
elif marks>=75:
    print("Grade B")
else:
    print("Grade C")
```

## Loops

```python
for i in range(5):
    print(i)

count=1
while count<=3:
    print("Count:",count)
    count+=1
```

## Functions

```python
def square(x):
    return x**2

print(square(5))

lambda_square=lambda x:x**2
print(lambda_square(4))
```

## Strings

```python
text='Data Science'
print(text.upper())
print(text.lower())
print(text.replace('Science','Analytics'))
print(text[:4])
```

## Lists

```python
fruits=['apple','banana','orange']
fruits.append('mango')
print(fruits)
print(fruits[0])
```

## Tuples

```python
coordinates=(10,20)
print(coordinates)
print(coordinates[0])
```

## Sets

```python
numbers={1,2,3,3,4}
print(numbers)
numbers.add(5)
print(numbers)
```

## Dictionaries

```python
student={'name':'Vignesh','age':21,'cgpa':7.8}
print(student['name'])
print(student.keys())
```

## List Comprehensions

```python
squares=[x**2 for x in range(10)]
print(squares)

even=[x for x in range(20) if x%2==0]
print(even)
```

## File Handling

```python
with open('sample.txt','w') as f:
    f.write('Hello Python')

with open('sample.txt','r') as f:
    print(f.read())
```

## Exception Handling

```python
try:
    result=10/0
except ZeroDivisionError:
    print('Cannot divide by zero')
finally:
    print('Execution Completed')
```

## Object Oriented Programming (OOP)

```python
class Student:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def display(self):
        print(self.name,self.age)

s1=Student('Vignesh',21)
s1.display()
```

## Student Grade Calculator

```python
marks=[85,90,78,92,88]
avg=sum(marks)/len(marks)
print('Average:',avg)
if avg>=90:
    print('Grade A')
elif avg>=80:
    print('Grade B')
else:
    print('Grade C')
```
