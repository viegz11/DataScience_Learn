# Matplotlib Basics

## Import Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plot

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.show()
```

## Bar Chart

```python
categories=['A','B','C']
values=[10,20,15]
plt.bar(categories,values)
plt.show()
```

## Histogram

```python
data=np.random.randn(1000)
plt.hist(data,bins=30)
plt.show()
```

## Scatter Plot

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y)
plt.show()
```

## Pie Chart

```python
sizes=[40,30,20,10]
plt.pie(sizes,labels=['A','B','C','D'])
plt.show()
```

## Subplots

```python
fig,ax=plt.subplots(1,2)
ax[0].plot([1,2,3])
ax[1].bar(['A','B'],[1,2])
plt.show()
```
