```python
y=data1['Exited'] # copied to check
X=data1.drop('Exited',axis=1)# removed for model
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=0) #0.75 data will be for training.
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,y_train) #command for training / fitting the model
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
