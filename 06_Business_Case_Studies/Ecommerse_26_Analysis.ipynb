```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.bar(df["Day_Of_Week"],df["Profit_Amount"])
plt.xlabel("Day of Week")
plt.ylabel("Profit Amount")
plt.title("Day of Week vs Profit Amount")
plt.show()
```

```python
sns.barplot(x="Month",y="Profit_Amount",data=df)
plt.show()
```

```python
sns.lineplot(x="Customer_Age",y="Quantity",data=df)
plt.show()
```

```python
gender_counts = df['Customer_Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Customer Gender')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
sns.barplot(x="Quantity",y="City",data=df)
plt.show()
```

```python
pd.crosstab(df['Membership_Status'], df['Customer_Gender']).plot(
    kind='bar',
    # stacked=True,
    figsize=(7,5)
)

plt.title("Customer Gender by Membership Status")
plt.xlabel("Membership Status")
plt.ylabel("Count")
plt.legend(title="Gender")
plt.show()
```

```python
sns.histplot(x='Delivery_Days',data=df)
plt.show()
```
